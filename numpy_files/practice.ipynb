```python
# NumPy is a powerful library for numerical operations in Python.
import numpy as np
```

```python
# Create a NumPy array 'a' with numbers from 1 to 10.
a = np.arange(1,11)
print(a)
```

```python
# Create a NumPy array 'b' with numbers from 11 to 20.
b = np.arange(11,21)
print(b)
```

```python
# Concatenate arrays 'a' and 'b' into a single array 'ab'.
ab = np.concatenate((a,b))
print(ab)
```

```python
# Calculate the product of all elements in 'ab' and round the result.
ab.prod().round()
```

```python
# Perform a dot product between arrays 'a' and 'b'.
a@b
```

```python
# Access the element at index 10 in the array 'ab'.
ab[10]
```

```python
# Create a new NumPy array 'ba'.
ba = np.array([9,8,7,6,5,4,3,2,6,4,8,3,54,677,867])
# Sort the elements of 'ba' in ascending order.
print(np.sort(ba))
```

```python
# Create a 3x3 NumPy array 'sm'.
sm = np.array([[1,2,3],
                [6,4,2],
                [0,9,7]])
# Sort the elements of each row (axis=1) of 'sm'.
ma = np.sort(sm, axis=1)
ma1 = np.sort(sm,axis=0)
print(ma)
```

```python
ma1
```

```python
# Sort the elements of each column (axis=0) of 'sm'.
ms = np.sort(sm, axis=0)
print(ms)
```

```python
# Square each element of the array 'sm'.
x = sm**2
print(x)
```

```python
# Create a new array 'sm' with numbers from 1 to 19.
sm = np.arange(1,20)
print(sm)
```

```python
# Create a list of indices.
indies = [2,3,0]
# Create a new array 'new' by selecting elements from 'sm' at the specified indices.
new = sm[indies]
print(new)
```

```python
# Create another list of indices.
index = [9,14,16]
# Create another new array 'new1' from 'sm' using the new indices.
new1 = sm[index]
print(new1)
```

```python
# Concatenate 'new' and 'new1'.
n = np.concatenate((new,new1))
# Reshape the concatenated array into a 2x3 matrix.
re = n.reshape(2,3)
print(re)
```

```python
# Transpose the matrix 're'.
re.T
```

```python
# Calculate the product of all elements in 're'.
re.prod()
```

```python
# Calculate the cumulative sum of the elements in 're'.
re.cumsum()
```

```python
# Flatten the matrix 're' into a 1D array.
re.flatten()
```

```python
# Print the 're' matrix.
print(re)
```

```python
# Create a 2x4 NumPy array 'y'.
y = np.array([[1,2,3,4],[5,6,7,8]])
print(y)
```

```python
# Create a 1x4 NumPy array 'z'.
z = np.array([[1,4,6,9]])
print(z)
```

```python
# Add arrays 'y' and 'z'. This works because of broadcasting.
yz = y+z
print(yz)
```

```python
f = np.vstack((y,z))
print(f)
```

```python
r = np.array([[1,2,3,4],[5,6,7,8]])
p = np.array([[10,20,30,40],[50,60,70,80]])
rp = np.hstack((r,p))
print(rp)
print()
print(np.vstack((r,p)))
```
