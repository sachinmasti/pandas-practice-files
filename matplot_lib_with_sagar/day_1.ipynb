```python
import matplotlib.pyplot as  plt
```

```python
x=[1,2,3,4]
y=[12,13,56,89]
plt.plot(x,y,color='black',linestyle ='--',marker='o',linewidth=3,label='sachin')
plt.title('sachin masti')
plt.legend(loc='upper left')
plt.grid(True,color='green')
plt.xlabel('sales',color='blue')
plt.ylabel('months',color='brown')
# plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['jan','feb','march','april'])
plt.yticks([1,2,3,4],['jan','feb','march','april'])
plt.show()
```

```python
product =['bat','ball','stumps','pads']
sales =[3000,1000,1500,1700]
regions =['east','west','north','south']
plt.barh(product,sales,color=['green','pink','yellow','red'])
# plt.legend(loc='upper right')
plt.xlabel('sales',color='brown')
plt.ylabel('products',color='blue')
plt.grid(True,color='pink')
plt.show()
```

```python
plt.pie(sales,labels=regions,autopct='%1.1f%%',colors=['green','gold','pink','yellow'])
plt.title('revenue by region')
plt.show()
```

```python
scores =[54,56,26,35,46,57,92,74,58,64,94,37,48,26]
plt.hist(scores,bins=5,color='purple',edgecolor='black')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Histogram of Scores')
plt.show()
```
