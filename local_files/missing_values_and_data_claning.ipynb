```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'Age': [22, 28, 35, 19]})

# If Age is 25 or more, assign 'Adult', otherwise 'Young'
df['Group'] = np.where(df['Age'] >= 25, 'Adult', 'Young')

#  Clean code
#  Super fast processing (thanks to NumPy!)
#  Perfect for creating new columns based on conditions


# Now we only want the rows where Age is 25 or more!
adults = df.query('Age >= 25')
print(adults)
#  Doesn't that look clean and elegant?
```
