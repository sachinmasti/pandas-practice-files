**English:** Import pandas and numpy, then create a DataFrame with several missing values (`np.nan`).
**Hindi:** Pandas aur numpy import karein, fir kai missing values (`np.nan`) ke saath ek DataFrame banayein.

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Amit', np.nan, 'Ramesh', 'Priya', 'Sanket', 'Anita', 'Rahul'],
    'Age': [23, 29, 35, 22, np.nan, 19, 29],
    'Department': ['Sales','HR','IT','Sales','IT','HR','Sales'],
    'Salary': [35000, 42000, 80000, 30000, 95000, 20000, np.nan],
    'Experience': [np.nan, 4, 10, 1, 15, 0.5, 5],
    'Performance_Score': [70, 88, 92, np.nan, 95, 55, 80]
}
df = pd.DataFrame(data)
```

**English:** Display the DataFrame to see its initial state.
**Hindi:** DataFrame ko uski shuruaati state dekhne ke liye display karein.

```python
df
```

**English:** Check for the total count of missing values in each column.
**Hindi:** Har column mein missing values ki total sankhya check karein.

```python
df.isna().sum()
```

**English:** Manually fill the missing 'Name' at index 1 with 'Neha'.
**Hindi:** Index 1 par missing 'Name' ko 'Neha' se manually fill karein.

```python
df.loc[1,'Name'] = 'Neha'
```

**English:** Fill the remaining missing values using the mean for numeric columns.
**Hindi:** Bachi hui missing values ko numeric columns ke liye mean ka istemal karke bharein.

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df['Experience'].fillna(df['Experience'].mean(),inplace = True)
df['Performance_Score'].fillna(df['Performance_Score'].mean(),inplace = True)
df['Age'].fillna(df['Age'].mean(),inplace = True)
```

**English:** Convert the data types of several columns to integer for cleaner data.
**Hindi:** Saaf data ke liye kai columns ke data types ko integer mein convert karein.

```python
df['Experience'] = df['Experience'].astype(int)
df['Salary'] = df['Salary'].astype(int)
df['Age'] = df['Age'].astype(int)
df['Performance_Score'] = df['Performance_Score'].astype(int)
```

**English:** Verify that all missing values have been filled.
**Hindi:** Verify karein ki sabhi missing values bhar di gayi hain.

```python
df.isna().sum()
```

**English:** Display the cleaned DataFrame.
**Hindi:** Cleaned DataFrame ko display karein.

```python
df
```

**English:** Group by 'Department' and 'Name' and perform multiple aggregations. Then, filter to show only the 'HR' department and sort by average salary.
**Hindi:** 'Department' aur 'Name' se group karein aur multiple aggregations perform karein. Fir, sirf 'HR' department dikhane ke liye filter karein aur average salary se sort karein.

```python
df.groupby(['Department','Name']).agg(
    avg_salary =('Salary','mean'),
    max_salary =('Salary','max'),
    min_salary =('Salary','min')
).round().query('Department == "HR"').sort_values(by='avg_salary',ascending=True)
```

**English:** Filter the DataFrame for employees older than 25, with more than 5 years of experience, and a performance score greater than 80.
**Hindi:** 25 se zyada umar, 5 saal se zyada experience, aur 80 se zyada performance score wale employees ke liye DataFrame filter karein.

```python
df[(df['Age'] > 25) & (df['Experience'] > 5)& (df['Performance_Score']> 80)]
```
