```python
# Import the pandas library, which is essential for data manipulation and analysis.
import pandas as pd
```

```python
# Load data from a CSV file into a pandas DataFrame.
df = pd.read_csv('new_csv.csv')
```

```python
# Get the dimensions (rows, columns) of the DataFrame.
df.shape
```

```python
# Display the first 5 rows of the DataFrame to get a quick overview of the data.
df.head()
```

```python
# Generate descriptive statistics for the numerical columns in the DataFrame.
df.describe()
```

```python
# Get a concise summary of the DataFrame, including data types and non-null values.
df.info()
```

```python
# Get a list of all column names in the DataFrame.
df.columns
```

```python
# Remove the 'Unnamed: 0' column from the DataFrame.
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Verify the new shape of the DataFrame after dropping the column.
df.shape
```

```python
# Display the updated list of column names.
df.columns
```

```python
# Select and display a subset of columns from the DataFrame.
df[['customer_id','city','dish_name']]
```

```python
# Select rows 0-10 and columns at index 3, 8, and 11 using iloc.
df.iloc[0:11,[3,8,11]]
```

```python
# Select rows 0-11 and specific columns by name using loc.
df.loc[0:11,['city','dish_name','price']]
```

```python
# Access a single value by label using .at.
df.at[2,'city']
```

```python
# Access a single value by integer position using .iat.
df.iat[2,3]
```

```python
# Create a new column 'new_age' by mapping values from the 'age' column.
new_age = {'Adult':25,'Senior':60,'Teenage':15}
df['new_age'] = df['age'].map(new_age)
```

```python
# Import the numpy library for numerical operations.
import numpy as np
```

```python
# Convert the 'new_age' column to a NumPy array with a specific data type.
np.array(df['new_age'], dtype = 'int32')
```

```python
# Filter the DataFrame for rows where 'new_age' is greater than 18 and 'rating' is greater than 3.
df[(df['new_age'] > 18)  & (df['rating']> 3)]
```

```python
# Use the query() method to perform the same filtering operation.
df.query('new_age > 20 and rating >3')
```

```python
# Check for and count the number of missing values in the 'new_age' column.
df['new_age'].isna().sum()
```

```python
# Count the number of times 'Cash' appears in the 'payment_method' column.
(df['payment_method'] == 'Cash').sum()
```

```python
# Count the number of times 'KFC' appears in the 'restaurant_name' column.
(df['restaurant_name'] == 'KFC').sum()
```
