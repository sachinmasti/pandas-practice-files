```python
import pandas as pd
```

```python
# import pandas as pd

data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'IT', 'Sales', 'Sales', 'Sales'],
    'Team': ['A', 'B', 'A', 'A', 'B', 'A', 'B', 'B'],
    'Employee': ['Amit', 'Priya', 'Raj', 'Neha', 'Ravi', 'Sonia', 'Vikram', 'Anil'],
    'Salary': [50000, 52000, 60000, 58000, 62000, 45000, 47000, 48000],
    'Experience': [3, 4, 5, 6, 7, 2, 3, 4]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.groupby(['Department','Team'])[['Salary','Experience']].mean().round().sort_values(by = 'Salary',ascending = True)
```

```python
df.groupby('Department').agg(
                    min_salary = ('Salary','min'),
                    mix_salary = ('Salary','max'),
                    avg_experience = ('Experience','mean')
)
```

```python
df.groupby('Department').filter(lambda y: y['Salary'].mean() > 55000)
```

```python
df.groupby('Department').filter(lambda y: y['Salary'].mean() > 55000)['Department']
```

```python
df['Salary_Rank'] = df.groupby('Department')['Salary'] .transform(lambda x: x.rank(ascending=False))
```

```python
df
```

```python
df.sort_values(['Department', 'Salary_Rank'])
```

```python
df.pivot_table(
    index='Department',
    values=['Salary','Experience'],
    aggfunc=(['sum','mean'])
)
```

```python
pivot_df = df.pivot_table(
    index='Department',
    values=['Salary','Experience'],
    aggfunc={'Salary':'sum', 'Experience':'mean'}
).rename(columns={'Salary':'Total_Salary','Experience':'Avg_Experience'})
```
